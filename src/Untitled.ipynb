```python
import math
from numba import njit
import numpy as np

@njit
def logFact(x):
    log_number = 0
    for i in range(1,x+1):
        log_number += math.log(i)
    return log_number

@njit
def multinomialOld(k0, k1, k2, n, p0, p1,p2):
    return math.exp(logFact(n)-(logFact(k1)+logFact(k2)+logFact(k0))) * p1**k1 * p2**k2 * (p0)**(k0)
@njit
def multinomial(k, n, p):
    coeff = logFact(n)
    rest = 1
    for i in range(len(k)):
        coeff -= logFact(k[i])
        rest *= p[i]**k[i]
    return math.exp(coeff) * rest

print(multinomialOld(970,10,20,1000,0.98,0.005,0.015),multinomial(k=np.array([970,10,20]),n=1000,p=np.array([0.98,0.005,0.015])))
```
